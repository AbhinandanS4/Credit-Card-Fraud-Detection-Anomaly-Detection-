```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.ensemble import IsolationForest
from sklearn.neighbors import LocalOutlierFactor
```

```python
# Calculate the correlation matrix
correlation_matrix = df.corr()

# Get the absolute correlation with the target variable (replace 'target_variable' with your target column name)
target_correlation = correlation_matrix['Class'].abs().sort_values(ascending=False)

# Plot the feature importance based on correlation
plt.figure(figsize=(10, 6))
sns.barplot(x=target_correlation.values[1:], y=target_correlation.index[1:])  # Exclude the target itself
plt.title('Feature Importance based on Correlation with Target')
plt.xlabel('Correlation')
plt.ylabel('Feature')
plt.xticks(rotation=90)
plt.show()
```

```python
# Define the threshold for feature importance
importance_threshold = 0.15

# Get the features with importance less than the threshold
features_to_drop = target_correlation[target_correlation < importance_threshold].index.tolist()
```

```python
features_to_drop
```

```python

# Drop the features from the DataFrame
df = df.drop(columns=features_to_drop)

# Print the remaining features
print("Remaining features:", df.columns.tolist())
```

# Random UnderSampling


```python
from imblearn.under_sampling import RandomUnderSampler

X = df.drop('Class', axis=1)  # Features
y = df['Class']  # Target variable

# Create a RandomUnderSampler object
rus = RandomUnderSampler(random_state=42)

# Fit and apply the resampling method
X_resampled, y_resampled = rus.fit_resample(X, y)

# Create a new DataFrame with the resampled data
df_resampled = pd.concat([X_resampled, y_resampled], axis=1)

# Print the class distribution after resampling
print(df_resampled['Class'].value_counts())

# Now you can use df_resampled for your modeling
```

```python
df=df_resampled.drop('Class',axis=1)
```

```python
df
```

## Model-Isolation Forest

```python
clf=IsolationForest(n_estimators=100,max_samples='auto',contamination=float(0.1),random_state=42)
```

```python
clf.fit(df)
```

```python
pred=clf.predict(df)
```

```python
np.unique(pred,return_counts=True)
```

```python
df['pred']=pred
```
